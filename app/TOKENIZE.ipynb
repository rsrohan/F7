```
from keras.preprocessing.text import text_to_word_sequence
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import string
import pickle
# define the text
#ps = PorterStemmer()
with open('/content/drive/My Drive/Colab Notebooks/Project_sem5/TOKENIZED.pickle', 'wb') as handle:
  for i in df['firstComment']:
    #tokenize the text, lowering and filtering
    sample = []
    entity_pre = ['@','#']
    result = text_to_word_sequence(i, filters='!"$%&()*+,-./:;<=>?[\\]^_`{|}~\t\n')
    for j in result:
      if j[0] not in entity_pre:    #removing # and @
        sample.append(j)
    samp = [word for word in sample if word.isalpha()]
    #stop_words = set(stopwords.words('english'))
    #sampl = [w for w in samp if not w in stop_words]
    #samp = [ps.stem(w) for w in samp]
    if len(samp)==0:
      continue
    pickle.dump(samp, handle, protocol=pickle.HIGHEST_PROTOCOL)
res = []
with (open('/content/drive/My Drive/Colab Notebooks/Project_sem5/TOKENIZED.pickle', "rb")) as openfile:       #reading that file
    while True:
        try:
            res = pickle.load(openfile)
            print(res)
            print(len(res))
        except EOFError:
            break
```
